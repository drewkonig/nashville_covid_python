```python
#replace Nan in ZIP, City column
#df_violations

df_violations['Latitude'].fillna('0.000000', inplace = True) 
df_violations['Longitude'].fillna('0.000000', inplace = True)
df_violations['ZIP'] = df_violations['ZIP'].replace(['TN'],'99999')
df_violations['ZIP'].fillna('99999', inplace = True) 
df_violations.fillna('UNKNOWN', inplace=True)

# ********************** Include Only Cases which are not closed status **************
df_violations = df_violations[df_violations['Status']!='Closed']
df_violations['City'] = df_violations['City'].str.upper() 

df_violations = pd.read_csv("../data/hubNashville311.csv")
df_violations = df_violations[df_violations['Request Type']=='COVID-19'].reset_index(drop=True)
df_violations = df_violations[df_violations['Subrequest Type']=='COVID-19 Violations'].reset_index(drop=True)
df_violations = df_violations[['Request #','Request Type','Subrequest Type','Date / Time Opened','Address','City','Council District','ZIP','Latitude','Longitude']]
df_violations.columns = ['request_id','request_type','subrequest_type','date_opened','address','city','district','zip','lat','long']

df_violations['date'] = df_violations.date_opened.str.split().str[0]
df_violations['time'] = df_violations.date_opened.str.split().str[1]




zipcode_data = df_violations.groupby(['City','ZIP']).aggregate(np.mean)

#zipcode_data
    
    
df_violations['count'] = 1
#temp = df_violations.groupby(['ZIP']).sum()
#temp = df_violations.groupby(['City']).sum()
temp = df_violations.groupby(['City', 'ZIP']).sum()                          
    
    
temp.reset_index(inplace = True)
#temp = temp[['ZIP','count']]
temp = temp[['City', 'ZIP','count']]
#zipcode_data = pd.merge(zipcode_data, temp, on='ZIP')
#zipcode_data = pd.merge(zipcode_data, temp, on='City)
zipcode_data = pd.merge(zipcode_data, temp,  how='left', left_on=['City','ZIP'], right_on = ['City','ZIP'])

zipcode_data['countInt'] = zipcode_data['count_y'].astype(np.int64)
zipcode_data['countStr'] = zipcode_data['count_y'].astype(str)
zipcode_data['ZIPStr'] = zipcode_data['ZIP'].astype(str).replace('\.0', '', regex=True)
zipcode_data['CityStr'] = zipcode_data['City'].astype(str)

#zipcode_data
```
