```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
violations = pd.read_csv("../data/hubNashville__311__Service_Requests.csv")
```

```python
violations=violations[violations['Request Type']=='COVID-19'].reset_index(drop=True)
violations=violations[violations['Subrequest Type']=='COVID-19 Violations'].reset_index(drop=True)
```

```python
violations = violations[['Request #', 'Request Type', 'Subrequest Type', 'Date / Time Opened', 'Address', 'City', 'Council District', 'ZIP', 'Latitude', 'Longitude', 'Mapped Location']]
violations.columns = ['request_id','request_type','subrequest_type','date_opened','address','city','district','zip','lat','lng','mapped_location']
violations.head(5)
```

```python
violations['date'] = violations.date_opened.str.split().str[0]
violations['time'] = violations.date_opened.str.split().str[1]
violations['am_pm'] = violations.date_opened.str.split().str[2]
```

```python
violations.head(5)
```

```python
fig, ax = plt.subplots()
ax.plot.hist(violations.date,bins=20)
fig.autofmt_xdate()
```
